```python
# 5  Mostre a mesma mensagem anterior mas utilizando o 
# '.format()' e o 'f-strings' (mostrar duas mensagens, uma para cada metodologia).

nome = 'Juliano'
altura = 1.72
peso = 80

print('{} tem {} de altura e pesa {:05d} kilos'.format(nome, altura, peso))
```

```python
# # 2 - Crie um programa que controle reservas de um hotel e mostre um menu com 3
# opes ao usurio:
# 1 - Fazer reserva
# 2 - Encerrar reserva
# 3 - Sair do programa
# Caso o usurio escolha a opo 1, o sistema deve receber o nome do cliente e quantos
# dias ele ficar hospedado.
# Mostre uma previso de valor total a pagar, sabendo que a diria do hotel  de R$ 80,00.
# Caso opo 2, o sistema deve receber o CPF e a quantidade de dias que o cliente ficou
# hospedado e mostrar o total do valor a pagar.
# Caso opo 3, o sistema deve ser encerrado. Mostre uma mensagem na tela.
diaria = 80.00
while True:
    
    print("Sistema Carlos Hotel, por favor digite sua opo: \n")
    entrada = input("Opes:\n 1 - Fazer reserva \n 2 - Encerrar reserva \n 3 - Sair do programa\n ")

    if entrada == '1':
        print("Ola voc escolheu a opo 1- Fazer uma reserva\n")
        nome_cliente = input("Entre com seu nome para cadastro \n")
        diaria_cliente = input("e quantos dias ir ficar hospedado? ")
        print(f"Seu nome  {nome_cliente} e voc ficar hospedado por {diaria_cliente}")
    elif entrada == '2':
        print("Ola voc escolheu a opo 2 - Fechar a conta\n")   
        cpf_cliente = input("Entre com seu CPF\n")
        diaria_cliente = float(input("entre com a quantidade que voc ficou hospedado\n"))
        diaria_cliente_usada = float(diaria_cliente * diaria)
        print(f"O cliente com CPF {cpf_cliente} tem uma diaria com valor de R${diaria_cliente_usada}\n")
    elif entrada == '3':
        print("Voc escolheu fechar o sistema...")
        break


            


```
