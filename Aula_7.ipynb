```python
# 5 – Mostre a mesma mensagem anterior mas utilizando o 
# '.format()' e o 'f-strings' (mostrar duas mensagens, uma para cada metodologia).

nome = 'Juliano'
altura = 1.72
peso = 80

print('{} tem {} de altura e pesa {:05d} kilos'.format(nome, altura, peso))
```

```python
# # 2 - Crie um programa que controle reservas de um hotel e mostre um menu com 3
# opções ao usuário:
# 1 - Fazer reserva
# 2 - Encerrar reserva
# 3 - Sair do programa
# Caso o usuário escolha a opção 1, o sistema deve receber o nome do cliente e quantos
# dias ele ficará hospedado.
# Mostre uma previsão de valor total a pagar, sabendo que a diária do hotel é de R$ 80,00.
# Caso opção 2, o sistema deve receber o CPF e a quantidade de dias que o cliente ficou
# hospedado e mostrar o total do valor a pagar.
# Caso opção 3, o sistema deve ser encerrado. Mostre uma mensagem na tela.
diaria = 80.00
while True:
    
    print("Sistema Carlos Hotel, por favor digite sua opção: \n")
    entrada = input("Opções:\n 1 - Fazer reserva \n 2 - Encerrar reserva \n 3 - Sair do programa\n ")

    if entrada == '1':
        print("Ola você escolheu a opção 1- Fazer uma reserva\n")
        nome_cliente = input("Entre com seu nome para cadastro \n")
        diaria_cliente = input("e quantos dias irá ficar hospedado? ")
        print(f"Seu nome é {nome_cliente} e você ficar hospedado por {diaria_cliente}")
    elif entrada == '2':
        print("Ola você escolheu a opção 2 - Fechar a conta\n")   
        cpf_cliente = input("Entre com seu CPF\n")
        diaria_cliente = float(input("entre com a quantidade que você ficou hospedado\n"))
        diaria_cliente_usada = float(diaria_cliente * diaria)
        print(f"O cliente com CPF {cpf_cliente} tem uma diaria com valor de R${diaria_cliente_usada}\n")
    elif entrada == '3':
        print("Você escolheu fechar o sistema...")
        break


            


```
