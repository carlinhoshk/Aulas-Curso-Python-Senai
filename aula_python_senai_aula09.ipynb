"""Faça uma lista de comprar com listas
O usuário deve ter a possibilidade de
inserir, apagar e listar valores da sua lista
Não permita que o programa quebre com 
erros de índices inexistentes na lista.
"""

```python
lista = ['chocolate', 'mesa', 'lousa']

while True:

    inserir_intem = input("Digite um intem para a lista: ")
    lista.append(inserir_intem)

    remove_intem = int(input("Digite um indicie da lista: "))
    lista.pop(remove_intem)

    lista_intem = input("Você deseja ver a lista? [s] ou [n]: ")
    if lista_intem == 's':
            for indicie, objeto in enumerate(lista):
                print(indicie, objeto)
    else:
        print("Você saiu")
        break            
```
