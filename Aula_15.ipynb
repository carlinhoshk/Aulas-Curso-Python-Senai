```python
with open("arq.txt", "w") as arquivo:
  arquivo.write("Usando o comando with para manipular arquivo!")
  arquivo.write("\nShow de bola!")
```

```python
with open("arq.txt","r") as arquivo:
  for i in arquivo.readlines():
    print(i)
```

```python
with open("numeros.txt", "w") as numeros:
  for num in range(1, 101):
    numeros.write(f'{num}\n')

#---------------------------------------------------------

with open("numeros.txt", "r") as origem:
  with open("impares.txt","a") as destino:
    for i in origem.readlines():
      if int(i)%2 !=0:
        destino.write(f'{i}\n')
```
