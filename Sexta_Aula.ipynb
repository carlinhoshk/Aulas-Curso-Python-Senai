```python
def isFloat(numero):
    try:
        float(numero)
        return True
    except:
        return False
        
while True:    
    numeroUm = input("Digite primeiro numero: ")
    numeroDois = input("Digite segundo numero: ")

    operacao = input("Digite operação a ser feita")


    numero_Um_float = isFloat(numeroUm)
    numero_Dois_float = isFloat(numeroDois)

    if numero_Um_float and numero_Dois_float:
        numero_float_um = float(numeroUm)
        numero_float_dois = float(numeroDois)
        if operacao == '+':
            calcula = float(numero_float_um + numero_float_dois)
            print(f"O resultado de {numero_float_um} mais {numero_float_dois} é igual a: {calcula}")
        elif operacao == '-':
            calcula = float(numero_float_um - numero_float_dois)
            print(f"O resultado de {numero_float_um} menos {numero_float_dois} é igual a: {calcula}")
        elif operacao == '*':
            calcula = float(numero_float_um * numero_float_dois)
            print(f"O resultado de {numero_float_um} vezes {numero_float_dois} é igual a: {calcula}")
        elif operacao == '/':
            calcula = float(numero_float_um / numero_float_dois)
            print(f"O resultado de {numero_float_um} dividido {numero_float_dois} é igual a: {calcula}")     
        else:
            print("Operação ou numero invalido.")    


    sair = input('Para sair digite sair')

    if sair == 'sair':
        break        
```
